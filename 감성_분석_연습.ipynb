```
from konlpy.tag import Okt

okt = Okt()

print(okt.pos('이 밤 그날의 반딧불을 당신의 창 가까이 보낼게요',norm=True, stem=True))
```

```
def tokenize(doc):
    # norm은 정규화, stem은 근어로 표시하기를 나타냄
    return ['/'.join(t) for t in okt.pos(doc,norm=True, stem=True)]
```

```
train_docs = [(tokenize(row[1]), row[2]) for row in train_data]
```

```
train_docs
```

```
test_docs = [(tokenize(row[1]), row[2]) for row in test_data]
```

```
test_docs
```

```
tokens = [t for d in train_docs for t in d[0]]
```

```
import nltk
text = nltk.Text(tokens, name='NMSC')
print(text)
```

```
# 전체 토큰의 개수
print(len(text.tokens))

# 중복을 제외한 토큰의 개수
print(len(set(text.tokens)))            

# 출현 빈도가 높은 상위 토큰 10개
print(text.vocab().most_common(10))
```

```
for f in text.vocab().most_common(200):
  print(f)
```

```
for f in text.vocab().most_common(200):
  print(f[0])
```

```
selected_words = [f[0] for f in text.vocab().most_common(200)]
```

```
def term_frequency(doc):
    return [doc.count(word) for word in selected_words]
```

```
for d,_ in train_docs:
  for word in selected_words:
    print(d.count(word))
```

```
# 각 문장 속 단어의 빈도수
train_x = [term_frequency(d) for d, _ in train_docs]
test_x = [term_frequency(d) for d, _ in test_docs]
```

```
train_x
```

```
# 각 문장 속 단어의 긍정 부정 라벨 
train_y = [c for _, c in train_docs]
test_y = [c for _, c in test_docs]
```

```
train_y
```

```
# 실수로 변환
import numpy as np

# 데이터의 특성
# 실수화 시킴
x_train = np.asarray(train_x).astype('float32')
x_test = np.asarray(test_x).astype('float32')

# 데이터의 라벨
# 실수화 시킴
y_train = np.asarray(train_y).astype('float32')
y_test = np.asarray(test_y).astype('float32')
```

```
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras import optimizers
from tensorflow.keras import losses
from tensorflow.keras import metrics

# input_shape=(200,) 우리는 200개의 상위단어만 체크했으니깐
model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(200,)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer=optimizers.RMSprop(lr=0.001),
             loss=losses.binary_crossentropy,
             metrics=[metrics.binary_accuracy])

model.fit(x_train, y_train, epochs=10, batch_size=512)
results = model.evaluate(x_test, y_test)
```

```
# 0.75의 정확도
results
```

```
def predict_pos_neg(review):
    token = tokenize(review)
    tf = term_frequency(token)
    data = np.expand_dims(np.asarray(tf).astype('float32'), axis=0)
    score = float(model.predict(data))
    if(score > 0.5):
        print("[{}]는 {:.2f}% 확률로 긍정 리뷰이지 않을까 추측해봅니다.^^\n".format(review, score * 100))
    else:
        print("[{}]는 {:.2f}% 확률로 부정 리뷰이지 않을까 추측해봅니다.^^;\n".format(review, (1 - score) * 100))
```

1. 문장을 받아온다.
2. 받아온 문장을 토큰화 한다.
3. 받아온 문장의 빈도수를 체크한다.
- 여기서 불용어를 처리하는 것이 핵심
4. 모델에 넣기 위해 실수화 시킨다.
5. 모델에 넣는다.
- 모델 안에서는 각 단어 마다 빈도가 있고 그 빈도의 값들이 부정인지 긍정인지 판단한다. 그리고 그 값들의 복합 여부에 따라서 긍정 부정은 판단한다.
6. 문장이 긍정인지 부정인지 반환한다.

```
# 학습 데이터의 수가 적어서 매우 낮은 정확도를 가짐 
predict_pos_neg("올해 최고의 영화! 세 번 넘게 봐도 질리지가 않네요.")
predict_pos_neg("배경 음악이 영화의 분위기랑 너무 안 맞았습니다. 몰입에 방해가 됩니다.")
predict_pos_neg("주연 배우가 신인인데 연기를 진짜 잘 하네요. 몰입감 ㅎㄷㄷ")
predict_pos_neg("믿고 보는 감독이지만 이번에는 아니네요")
predict_pos_neg("주연배우 때문에 봤어요")
predict_pos_neg('아깝다 욕나온다')
```
